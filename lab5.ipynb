```python
class MyAnimals:
    def __init__(self, animals, name, age, mass=None) -> None:
        self.animals = animals
        self.name = name
        self.age = age
        self.mass = 0 if mass is None else mass

    def speak(self):
        if self.animals == "Кіт":
            return f"Моя Тваринка {self.animals} говорить Мяууу"
        elif self.animals == "Собака":
            return f"Моя Тваринка {self.animals} говорить Гавв"
        else:
            return f"Моя Тваринка {self.animals} не розмовляє"

    def add_one_year(self):
        self.age += 1
        return f"Моя Тваринка {self.animals} прожила зі мною один рік. Тепер їй {self.age}"

a1 = MyAnimals("Рибки", "Неони", 1)
a2 = MyAnimals("Кіт", "Мурка", 5, 10)
a3 = MyAnimals("Собака", "Рекс", 2)

# Застосовуємо методи до всіх об'єктів
objects = [a1, a2, a3]
for obj in objects:
    print(f"--- {obj.name} ---")
    print(obj.speak())
    print(obj.add_one_year())
```

```python
class MyAnimals:
    total_animals = 0
    total_animal_types = set()

    def __init__(self, animals, name, age) -> None:
        self.animals = animals
        self.name = name
        self.age = age
        MyAnimals.total_animal_types.add(animals)
        MyAnimals.total_animals += 1

# Створення об'єктів
a1 = MyAnimals("Рибки", "Неони", 1)
MyAnimals("Рибки", "Гупії", 1)
MyAnimals("Кіт", "Мурка", 5)
MyAnimals("Собака", "Рекс", 2)

print(f"Всі види тварин: {list(MyAnimals.total_animal_types)}")
print(f"Загальна кількість: {MyAnimals.total_animals}")

# Демонстрація різниці доступу
print(f"Доступ з класу: {MyAnimals.total_animal_types}")
a1.total_animal_types = {"Тільки Риби"} # Перекриття атрибуту
print(f"Доступ з a1 (локально змінений): {a1.total_animal_types}")
print(f"Доступ з класу (незмінний): {MyAnimals.total_animal_types}")
```

```python
from datetime import datetime

class MyAnimals:
    max_age = 20
    def __init__(self, animals, name, age) -> None:
        self.animals = animals
        self.name = name
        self.age = age

    @property
    def live_untill(self):
        return datetime.today().year - self.age + self.max_age

    # --- ЗАВДАННЯ: Додана нова властивість ---
    @property
    def born_year(self):
        return datetime.today().year - self.age

a1 = MyAnimals("Кіт", "Мурка", 5)
print(f"Мій {a1.animals} буде зі мною до {a1.live_untill} року.")
print(f"Рік народження тваринки: {a1.born_year}")
```

```python
class MyAnimals:
    total_objects = 0

    def __init__(self, animals, name, age) -> None:
        self.animals = animals
        self.name = name
        self.age = age
        MyAnimals.total_objects += 1

    @classmethod
    def from_csv(cls, csv_data: str):
        animal, name, age = csv_data.split(",")
        # Перевірка валідації, як просили
        if not cls.check_name_capitalized(name):
            print(f"Увага! Ім'я '{name}' виправлено на '{name.capitalize()}'")
            name = name.capitalize()
        return cls(animal, name, int(age))

    # --- ЗАВДАННЯ: Метод перевірки великої літери ---
    @classmethod
    def check_name_capitalized(cls, name):
        return name[0].isupper()

# Тестуємо
a1 = MyAnimals.from_csv("Кіт,Мурка,5")
a2 = MyAnimals.from_csv("Собака,бобік,3")

print(f"Тваринка 1: {a1.name}")
print(f"Тваринка 2: {a2.name}")
```

```python
class MyAnimals:
    def __init__(self, animals, name) -> None:
        self.animals = animals
        self.name = name
        self._secret_phrase = "Секретна кличка (protected)"
        self.__secret_strong = "Пароль від сейфа (private)"

    @property
    def access_private(self):
        return f"Доступ через властивість: {self.__secret_strong}"

a1 = MyAnimals("Собака", "Рекс")

print(a1._secret_phrase) # Можна, але не рекомендується
# print(a1.__secret_strong) # Це викличе помилку AttributeError!

# Як обійти захист (name mangling):
print(f"Злом приватного атрибуту: {a1._MyAnimals__secret_strong}")

# Правильний доступ:
print(a1.access_private)
```

```python
class MyAnimals:
    def __init__(self, animals, name, age) -> None:
        self.animals = animals
        self.name = name
        self.age = age

    def __repr__(self) -> str:
        return f'MyAnimals("{self.animals}", "{self.name}", {self.age})'

    def __str__(self) -> str:
        return f"Тваринка: {self.animals} на ім'я {self.name}, вік: {self.age}"

a1 = MyAnimals("Собака", "Рекс", 2)
print("Вивід print (str):", a1)
print("Вивід repr:", repr(a1))
```

```python
class MyAnimals:
    def __init__(self, animals, name, age) -> None:
        self.animals = animals
        self.name = name
        self.age = age

    # Віднімання
    def __sub__(self, obj):
        if isinstance(obj, MyAnimals):
            return f"{self.name} покусав {obj.name}"
        elif isinstance(obj, int):
            self.age -= obj # Віднімаємо роки
            return f"{self.name} помолодшав на {obj} років, тепер йому {self.age}"

    # Порівняння (більше)
    def __gt__(self, obj):
        if isinstance(obj, MyAnimals):
            return self.age > obj.age
        return False

    # Довжина
    def __len__(self) -> int:
        return len(self.name)

a = MyAnimals("Собака", "Рекс", 5)
b = MyAnimals("Кіт", "Мурзик", 3)

print(f"Порівняння віку (Рекс > Мурзик): {a > b}")
print(f"Віднімання об'єктів: {a - b}")
print(f"Віднімання числа: {a - 2}") # Рексу стане 3 роки
print(f"Довжина імені Рекс: {len(a)}")
```
